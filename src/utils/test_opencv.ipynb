```python
def col_extraction(image, pt, rayon):
    r = 0
    g = 0
    b = 0
    for x in range(max(pt.x - rayon,0), min(pt.x+rayon, images.shape[1])):
        for y in range(max(pt.y - rayon,0), min(pt.y+rayon, images.shape[0])):
            r += images[y][x][0]
            g += images[y][x][1]
            b += images[y][x][2]
    r /= (rayon*2)**2
    g /= (rayon*2)**2
    b /= (rayon*2)**2
    return (r,g,b)
```
